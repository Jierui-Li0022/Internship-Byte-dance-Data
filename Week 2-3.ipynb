# Part 2

# Task 1 : Import the dataset and perform training set test set segmentation on the dataset.

One-Hot Encoding

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# 
data = pd.read_csv(r"C:\Users\Jerry\Desktop\2024\Week 2-3\data_raw.csv")

# 
data_encoded = pd.get_dummies(data)

#  y  X
y = data_encoded["exp_sig"]
X = data_encoded.drop(columns=["exp_sig"], axis=1)

# 70% 30% 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Task 2 : Logistic regression, GBDT, mlp models are established separately.

Here, whether we predict or train the model, we are able to observe that this is a classification model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

Learn about the sklearn package and build logistic regression, GBDT and MLP models.

```python
# 
log_reg = LogisticRegression(max_iter=1000)
gbdt = GradientBoostingClassifier()
mlp = MLPClassifier(max_iter=1000)
```

# Task 3 : Using three models to train the training set.  
For each algorithm, the corresponding precision score, recall score, f1 score and other parameters are displayed.The testdata matrix corresponding to each of the three algorithms is output.

We use precision scorerecall scoref1 score these three parameters to as a significant metric for evaluating the performance of classification models.

# 

```python
# 
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Logistic Regression Report:\n", classification_report(y_test, y_pred_log_reg))
```

# GBDT

```python
# GBDT
gbdt = GradientBoostingClassifier()
gbdt.fit(X_train, y_train)
y_pred_gbdt = gbdt.predict(X_test)
print("GBDT Accuracy:", accuracy_score(y_test, y_pred_gbdt))
print("GBDT Report:\n", classification_report(y_test, y_pred_gbdt))
```

# MLP

```python
# MLP
mlp = MLPClassifier(max_iter=1000)
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
print("MLP Accuracy:", accuracy_score(y_test, y_pred_mlp))
print("MLP Report:\n", classification_report(y_test, y_pred_mlp))
```

# Task 4 : Adjustment of model hyperparameters.

Hyperparameters of a model are parameters that need to be set before training in a machine learning algorithm.  
Such as:   
1. Learning Rate
2. Regularization Parameters
3. Batch Size
4. Number of Epochs
5. Number of Hidden Layers and Neurons per Layer
6. Max Depth of Decision Tree

Here our approach to tuning hyperparameters is Grid Search, i.e., an exhaustive search of all possible values of each hyperparameter to find the optimal combination.

# Logistic Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# 
param_grid_log_reg = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear']
}

# GridSearchCV
grid_search_log_reg = GridSearchCV(LogisticRegression(max_iter=1000), param_grid_log_reg, cv=5, scoring='accuracy')
grid_search_log_reg.fit(X_train, y_train)

# 
print("Best parameters for Logistic Regression: ", grid_search_log_reg.best_params_)
# 
best_log_reg = grid_search_log_reg.best_estimator_

# 
y_pred_log_reg = best_log_reg.predict(X_test)
print("Best Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Best Logistic Regression Report:\n", classification_report(y_test, y_pred_log_reg))
```

# Gradient Boosting Decision Tree, GBDT

```python
from sklearn.ensemble import GradientBoostingClassifier

# 
param_grid_gbdt = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],  # 
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0]
}

# GridSearchCV
grid_search_gbdt = GridSearchCV(GradientBoostingClassifier(), param_grid_gbdt, cv=5, scoring='accuracy')
grid_search_gbdt.fit(X_train, y_train)

# 
print("Best parameters for GBDT: ", grid_search_gbdt.best_params_)
# 
best_gbdt = grid_search_gbdt.best_estimator_

# GBDT
y_pred_gbdt = best_gbdt.predict(X_test)
print("Best GBDT Accuracy:", accuracy_score(y_test, y_pred_gbdt))
print("Best GBDT Report:\n", classification_report(y_test, y_pred_gbdt))
```

# Multi-Layer Perceptron, MLP

```python
from sklearn.neural_network import MLPClassifier

# 
param_grid_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['tanh', 'relu'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001],
    'learning_rate': ['constant', 'adaptive'],
    'learning_rate_init': [0.001, 0.01, 0.1],  # 
    'batch_size': [32, 64, 128]  # 
}

# GridSearchCV
grid_search_mlp = GridSearchCV(MLPClassifier(max_iter=1000), param_grid_mlp, cv=5, scoring='accuracy')
grid_search_mlp.fit(X_train, y_train)

# 
print("Best parameters for MLP: ", grid_search_mlp.best_params_)
# 
best_mlp = grid_search_mlp.best_estimator_

# MLP
y_pred_mlp = best_mlp.predict(X_test)
print("Best MLP Accuracy:", accuracy_score(y_test, y_pred_mlp))
print("Best MLP Report:\n", classification_report(y_test, y_pred_mlp))
```

```python
import pickle

# 
with open('linear_regression_model.pkl', 'wb') as f:
    pickle.dump(log_reg, f)

# GBDT
with open('gbdt_model.pkl', 'wb') as f:
    pickle.dump(gbdt, f)

# MLP
with open('mlp_model.pkl', 'wb') as f:
    pickle.dump(mlp, f)
```

```python
from flask import Flask, request, jsonify, render_template
import pickle
import numpy as np

app = Flask(__name__)

# 
with open('linear_regression_model.pkl', 'rb') as f:
    linear_model = pickle.load(f)

with open('gbdt_model.pkl', 'rb') as f:
    gbdt_model = pickle.load(f)

with open('mlp_model.pkl', 'rb') as f:
    mlp_model = pickle.load(f)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    model_type = request.form['model_type']
    features = request.form['features']
    features = np.array([float(x) for x in features.split(',')]).reshape(1, -1)

    if model_type == 'linear':
        prediction = linear_model.predict(features)
    elif model_type == 'gbdt':
        prediction = gbdt_model.predict(features)
    elif model_type == 'mlp':
        prediction = mlp_model.predict(features)
    else:
        return jsonify({'error': ''})

    return render_template('index.html', prediction=prediction[0])

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle

app = Flask(__name__)
f = open(f"gbdt_model.pkl", "rb")
model = pickle.load(f)
f.close()

@app.route('/')
def home():
    return render_template('page.html')

@app.route('/predict', methods=['POST'])
def predict():
    features_list = list(request.form.values())
    features = np.array(features_list).reshape(1, -1)
    predict_outcome = model.predict(features)
    return render_template('page.html',
                           prediction_display_area='Predict Outcome{}'.format(predict_outcome)
                           )
if __name__ == "__main__":
    app.run(port=5700, debug=True)
```
