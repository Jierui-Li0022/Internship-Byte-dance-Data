# Graph 3 and 4

接下来画的是一副三维散点图，这幅三维散点图的目的是可视化核心影响量化（core_inf）、平均浏览量（avg_pv）、账号创作级别（cre_lev）以及粉丝互动（fan_cre）之间的关系。具体来说：  

1. X轴（平均浏览量 avg_pv）：代表每个账号的平均浏览量。
2. Y轴（账号创作级别 cre_lev）：代表每个账号的创作级别，使用标签编码处理后的数值。
3. Z轴（核心影响量化 core_inf）：代表每个账号的核心影响量化分数。
4. 点的颜色（粉丝互动 fan_cre）：颜色编码表示粉丝互动情况，红色表示有粉丝互动，蓝色表示没有粉丝互动。  

在这里，我们希望通过可视化和分析数据，目的是找出哪些因素影响核心影响量化（core_inf）。这是为了进一步理解哪些因素可能对核心影响量化有显著影响。

"核心影响量化"（core_inf）是一个指标，用来衡量一个主体（例如一个账号、内容创作者、用户等）在特定平台或领域中的影响力。这个指标通常会结合多个因素，如用户互动、粉丝数量、内容质量、转化率等，来综合评估其核心影响力。

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.font_manager import FontProperties
from sklearn.preprocessing import LabelEncoder

# 确保相关列为数值类型
df1['avg_pv'] = pd.to_numeric(df1['avg_pv'], errors='coerce')
df1['core_inf'] = pd.to_numeric(df1['core_inf'], errors='coerce')
# 使用标签编码处理分类数据
label_encoder = LabelEncoder()
df1['cre_lev'] = label_encoder.fit_transform(df1['cre_lev'])
df1['fan_cre'] = df1['fan_cre'].astype(int)  # 将布尔值转换为整数

# 设置中文字体
font_path = "C:\\Windows\\Fonts\\simhei.ttf"  # Windows系统中的SimHei（黑体）字体路径
font_prop = FontProperties(fname=font_path)

# 创建三维散点图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制散点
sc = ax.scatter(df1['avg_pv'], df1['cre_lev'], df1['core_inf'], c=df1['fan_cre'], cmap=cm.coolwarm, s=50)

# 添加颜色条
cbar = plt.colorbar(sc, ax=ax, pad=0.1)
cbar.set_label('粉丝互动', fontproperties=font_prop)

# 设置轴标签和标题
ax.set_xlabel('平均浏览量 (avg_pv)', fontproperties=font_prop)
ax.set_ylabel('账号创作级别 (cre_lev)', fontproperties=font_prop)
ax.set_zlabel('核心影响量化 (core_inf)', fontproperties=font_prop)
plt.title('三维图：核心影响量化、平均浏览量和账号创作级别与粉丝互动', fontproperties=font_prop)

# 显示图形
plt.show()
```

由于数据量过于庞大，为了使图表更容易观察，我们可以减少数据点的数量，这一点可以通过以下方法来实现：

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.font_manager import FontProperties
from sklearn.preprocessing import LabelEncoder

# 确保相关列为数值类型
df1['avg_pv'] = pd.to_numeric(df1['avg_pv'], errors='coerce')
df1['core_inf'] = pd.to_numeric(df1['core_inf'], errors='coerce')
# 使用标签编码处理分类数据
label_encoder = LabelEncoder()
df1['cre_lev'] = label_encoder.fit_transform(df1['cre_lev'])
df1['fan_cre'] = df1['fan_cre'].astype(int)  # 将布尔值转换为整数

# 随机采样，减少数据量
df_sampled = df1.sample(frac=0.05, random_state=1)  # 抽取5%的数据

# 设置中文字体
font_path = "C:\\Windows\\Fonts\\simhei.ttf"  # Windows系统中的SimHei（黑体）字体路径
font_prop = FontProperties(fname=font_path)

# 创建三维散点图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制散点
sc = ax.scatter(df_sampled['avg_pv'], df_sampled['cre_lev'], df_sampled['core_inf'], c=df_sampled['fan_cre'], cmap=cm.coolwarm, s=50)

# 添加颜色条
cbar = plt.colorbar(sc, ax=ax, pad=0.1)
cbar.set_label('粉丝互动', fontproperties=font_prop)

# 设置轴标签和标题
ax.set_xlabel('平均浏览量 (avg_pv)', fontproperties=font_prop)
ax.set_ylabel('账号创作级别 (cre_lev)', fontproperties=font_prop)
ax.set_zlabel('核心影响量化 (core_inf)', fontproperties=font_prop)
plt.title('三维图：核心影响量化、平均浏览量和账号创作级别与粉丝互动', fontproperties=font_prop)

# 显示图形
plt.show()
```

细分每个因素来讨论：    

1. 平均浏览量（avg_pv）与核心影响量化（core_inf）：的关系：
数据点在X轴（平均浏览量）和Z轴（核心影响量化）的分布相对均匀，说明在现有数据中，这两个变量之间没有明显的相关性。也就是说，平均浏览量的高低并不明显影响核心影响量化的分数。  

2. 账号创作级别（cre_lev）与核心影响量化（core_inf）：的关系：
数据点在Y轴（账号创作级别）和Z轴（核心影响量化）的分布也较为均匀，说明账号创作级别对核心影响量化的分数没有显著影响。不同创作级别的账号在核心影响量化上的表现并没有显著差异。

3. 粉丝互动的分布：
颜色表示粉丝互动情况，从图中可以看到红色和蓝色点混合在一起，表明粉丝互动与其他三个变量之间没有显著的分离或聚集现象。换句话说，粉丝互动情况在不同的平均浏览量、账号创作级别和核心影响量化分数下都是均匀分布的。

从这幅三维散点图中可以得出以下结论：  

在当前数据集下，平均浏览量、账号创作级别和核心影响量化之间没有显著的线性关系。粉丝互动的情况也没有明显的模式与上述三个变量关联。

# Task 1 : Import the dataset and perform training set test set segmentation on the dataset.

既然数据集包含了字母和数字，你需要对这些混合类型的数据进行预处理，将非数值数据转换为适合模型使用的数值数据。通常的做法是将分类数据转换为独热编码（One-Hot Encoding）

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# 导入数据集
data = pd.read_csv(r"C:\Users\Jerry\Desktop\2024暑实习\Week 2-3\data_raw.csv")

# 将分类数据转换为独热编码
data_encoded = pd.get_dummies(data)

# 提取目标变量 y 和特征变量 X
y = data_encoded["exp_sig"]
X = data_encoded.drop(columns=["exp_sig"], axis=1)

# 分割数据集：70% 用于训练，30% 用于测试
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Task 2 : Logistic regression, GBDT, mlp models are established separately.

Here, whether we predict or train the model, we are able to observe that this is a classification model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

Learn about the sklearn package and build logistic regression, GBDT and MLP models.

```python
# 建立模型
log_reg = LogisticRegression(max_iter=1000)
gbdt = GradientBoostingClassifier()
mlp = MLPClassifier(max_iter=1000)
```

# Task 3 : Using three models to train the training set.  
For each algorithm, the corresponding precision score, recall score, f1 score and other parameters are displayed.The testdata matrix corresponding to each of the three algorithms is output.

We use precision score、recall score、f1 score these three parameters to as a significant metric for evaluating the performance of classification models.

Precision（精确率）：精确率是指模型预测为正类的样本中实际为正类的比例.  
Recall（召回率）：召回率是指实际为正类的样本中被模型正确预测为正类的比例.  
F1 Score（F1分数）：F1分数是精确率和召回率的调和平均数，用于综合评估模型的性能.

# 逻辑回归模型：

```python
# 建立逻辑回归模型
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Logistic Regression Report:\n", classification_report(y_test, y_pred_log_reg))
```

# GBDT模型：

```python
# 建立GBDT模型
gbdt = GradientBoostingClassifier()
gbdt.fit(X_train, y_train)
y_pred_gbdt = gbdt.predict(X_test)
print("GBDT Accuracy:", accuracy_score(y_test, y_pred_gbdt))
print("GBDT Report:\n", classification_report(y_test, y_pred_gbdt))
```

# MLP模型：

```python
# 建立MLP模型
mlp = MLPClassifier(max_iter=1000)
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
print("MLP Accuracy:", accuracy_score(y_test, y_pred_mlp))
print("MLP Report:\n", classification_report(y_test, y_pred_mlp))
```

# Task 4 : Adjustment of model hyperparameters.

Hyperparameters of a model are parameters that need to be set before training in a machine learning algorithm.  
Such as:   
1. 学习率（Learning Rate）
2. 正则化参数（Regularization Parameters）
3. 批量大小（Batch Size）
4. 迭代次数（Number of Epochs）
5. 隐藏层数和每层神经元数（Number of Hidden Layers and Neurons per Layer）
6. 决策树的最大深度（Max Depth of Decision Tree）

Here our approach to tuning hyperparameters is Grid Search, i.e., an exhaustive search of all possible values of each hyperparameter to find the optimal combination.

# 逻辑回归（Logistic Regression）超参数调整：

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# 定义超参数网格
param_grid_log_reg = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear']
}

# 创建GridSearchCV对象
grid_search_log_reg = GridSearchCV(LogisticRegression(max_iter=1000), param_grid_log_reg, cv=5, scoring='accuracy')
grid_search_log_reg.fit(X_train, y_train)

# 最佳超参数
print("Best parameters for Logistic Regression: ", grid_search_log_reg.best_params_)
# 最佳模型
best_log_reg = grid_search_log_reg.best_estimator_

# 评估最佳模型
y_pred_log_reg = best_log_reg.predict(X_test)
print("Best Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Best Logistic Regression Report:\n", classification_report(y_test, y_pred_log_reg))
```

# 梯度提升决策树（Gradient Boosting Decision Tree, GBDT）超参数调整：

```python
from sklearn.ensemble import GradientBoostingClassifier

# 定义超参数网格
param_grid_gbdt = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],  # 学习率
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0]
}

# 创建GridSearchCV对象
grid_search_gbdt = GridSearchCV(GradientBoostingClassifier(), param_grid_gbdt, cv=5, scoring='accuracy')
grid_search_gbdt.fit(X_train, y_train)

# 最佳超参数
print("Best parameters for GBDT: ", grid_search_gbdt.best_params_)
# 最佳模型
best_gbdt = grid_search_gbdt.best_estimator_

# 评估最佳GBDT模型
y_pred_gbdt = best_gbdt.predict(X_test)
print("Best GBDT Accuracy:", accuracy_score(y_test, y_pred_gbdt))
print("Best GBDT Report:\n", classification_report(y_test, y_pred_gbdt))
```

# 多层感知器（Multi-Layer Perceptron, MLP）超参数调整：

```python
from sklearn.neural_network import MLPClassifier

# 定义超参数网格
param_grid_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['tanh', 'relu'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001],
    'learning_rate': ['constant', 'adaptive'],
    'learning_rate_init': [0.001, 0.01, 0.1],  # 学习率
    'batch_size': [32, 64, 128]  # 批量大小
}

# 创建GridSearchCV对象
grid_search_mlp = GridSearchCV(MLPClassifier(max_iter=1000), param_grid_mlp, cv=5, scoring='accuracy')
grid_search_mlp.fit(X_train, y_train)

# 最佳超参数
print("Best parameters for MLP: ", grid_search_mlp.best_params_)
# 最佳模型
best_mlp = grid_search_mlp.best_estimator_

# 评估最佳MLP模型
y_pred_mlp = best_mlp.predict(X_test)
print("Best MLP Accuracy:", accuracy_score(y_test, y_pred_mlp))
print("Best MLP Report:\n", classification_report(y_test, y_pred_mlp))
```

```python
from flask import Flask, request, jsonify, render_template
import pickle
import numpy as np

app = Flask(__name__)

# 加载训练好的模型
with open('linear_regression_model.pkl', 'rb') as f:
    linear_model = pickle.load(f)

with open('gbdt_model.pkl', 'rb') as f:
    gbdt_model = pickle.load(f)

with open('mlp_model.pkl', 'rb') as f:
    mlp_model = pickle.load(f)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    model_type = request.form['model_type']
    features = request.form['features']
    features = np.array([float(x) for x in features.split(',')]).reshape(1, -1)

    if model_type == 'linear':
        prediction = linear_model.predict(features)
    elif model_type == 'gbdt':
        prediction = gbdt_model.predict(features)
    elif model_type == 'mlp':
        prediction = mlp_model.predict(features)
    else:
        return jsonify({'error': '无效的模型类型'})

    return render_template('index.html', prediction=prediction[0])

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle

app = Flask(__name__)
f = open(f"gbdt_model.pkl", "rb")
model = pickle.load(f)
f.close()

@app.route('/')
def home():
    return render_template('page.html')

@app.route('/predict', methods=['POST'])
def predict():
    features_list = list(request.form.values())
    features = np.array(features_list).reshape(1, -1)
    predict_outcome = model.predict(features)
    return render_template('page.html',
                           prediction_display_area='Predict Outcome：{}'.format(predict_outcome)
                           )
if __name__ == "__main__":
    app.run(port=5700, debug=True)
```
